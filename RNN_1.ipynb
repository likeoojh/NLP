Recurrent Neural Networks

- The main advantages to use RNN for text data is that it reduces the number of parameters of model (by avoiding one-hot encoding) and it shares weights between different positions of the text
- RNNs model sequence data and can have different lengths of inputs and outputs.

Sequence to sequence models
- Many to one : classification tasks, where the final output is a probability distribution.
